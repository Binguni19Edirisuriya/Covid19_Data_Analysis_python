#### COncatenation and transformation

```python
covidUS = covid[covid['Country/Region'] == 'US']
covidAgf = covid[covid['Country/Region'] == 'Afghanistan']
covidUS
covidAgf

Covidconcat = pd.concat([covidUS,covidAgf])
Covidconcat
```

#### Transforming data using stack

```python
covidUS = covid[covid['Country/Region'] == 'US']
covidLatLong = covidUS.drop(['Lat','Long'],axis = 1)
covidTransform = covidLatLong.stack()
covidTransform = covidTransform.reset_index()
covidTransform
```

#### Renaming columns

```python
covidTransform = covidTransform[covidTransform['level_1'] != 'Country/Region']
covidTransformFilter = covidTransform[['level_1',0]]
covidTransformFilter.columns = ['Date','No of Cases']
covidTransformFilter = covidTransformFilter.sort_values(by = 'No of Cases',ascending = False)
covidTransformFilter
```

#### Sorting the data

```python
covidSort = covidTransformFilter.sort_values(by = 'No of Cases',ascending = False)
covidSort
```

#### Parsing data in Pandas

```python
covidSort['Date'] = pd.to_datetime(covidSort.Date)
covidSort[0:5]
```

### Group and Aggregation

```python
covidGroup = covidSort.groupby([pd.Grouper(key = 'Date' , freq = 'SM')])['No of Cases'].sum().reset_index().sort_values(by = 'Date', ascending = True)
covidGroup
```

### Taking substrings and parsing Date

```python
covidGroup['Date'] = covidGroup['Date'].astype(str)
covidGroup['Parsed Date'] = covidGroup['Date'].str[5:]
covidGroup['Parsed Date']=covidGroup['Parsed Date'].replace({'-':'/'},regex = True)
covidGroup
covidFinal = covidGroup[['No of Cases','Parsed Date']]
covidCases = covidFinal.sort_values(by='Parsed Date',ascending = True)
covidCases
```
