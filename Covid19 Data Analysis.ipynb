## Flitering and Selecting

```python
import pandas as pd
```

```python
covid_data_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
covid = pd.read_csv(covid_data_url)
covid.head()
```

```python
##Indexing is used to uniquely identify rows and columns
```

#### Selecting one field using pandas

```python
Covidlat = covid['Lat']
Covidlat
```

#### Selecting multiple fields using python

```python
Covidlatlong = covid[['Lat','Long']]
Covidlatlong
```

#### Selecting specific field with specific rows in pandas

```python
covidselect = covid.loc[[1,4],['Lat','Long']]
covidselect
```

#### Selecting range of rows 

```python
covidRange = covid.loc[1:4,['Lat','Long']]
covidRange
```

#### Apply filter for a specific field

```python
covidFilter = covid[covid['Lat']>31]
covidFilter
```

#### Dropping columns in pandas

```python
covidDrop = covid.drop(['Lat','Long'],axis = 1)
covidDrop
```
